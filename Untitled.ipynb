```python
# test
def getAllRow():
    listRow = driver.find_elements_by_css_selector('tbody tr')
    return listRow


def getListLink():
    listLink = []
    for i in getAllRow():
        listTD = i.find_elements_by_css_selector('td')
        listLink.append('https://drugbank.vn/thuoc/'+slug.slug(listTD[1].text)+'&'+listTD[0].text)
    return listLink

def getListLinkByPage(int):
    listLink = []
    for i in range(1,int):
        driver.get('https://drugbank.vn/danh-sach-thuoc?page='+str(i)+'&size=20&sort=tenThuoc,asc')
        for j in getListLink():
            listLink.append(j)
    return listLink
```

```python
with open('app.json') as json_file:
    data = json.load(json_file)
```

```python
def handleGia(moneyType):
    arr = moneyType
    return arr[1:]

def getInforOfThuoc():
    dic = {}
    temp = driver.find_elements_by_css_selector('dl dd')
    try:
        dic["DonViTinh"] = driver.find_element_by_css_selector('table.table.table-striped.table-narrow tbody tr').find_elements_by_css_selector('td')[4].text
        dic["Gia"] = handleGia(driver.find_element_by_css_selector('table.table.table-striped.table-narrow tbody tr').find_elements_by_css_selector('td')[3].text)
        dic["TenThuoc"] = temp[0].text
        dic["ThuocId"] = temp[1].text
        try:
            dic["MoTa"] = temp[3].text
        except:
            dic["MoTa"] = ""
        dic["HanSuDung"] = temp[6].text.split(" ")[0]
        try:
            dic["DangBaoChe"] = temp[4].text
        except:
            dic["DangBaoChe"] = ""
        try:
            dic["QuyCachDongGoi"] = temp[5].text
        except:
            dic["QuyCachDongGoi"] = ""
        try:
            dic["TenNhaCungCap"] = temp[7].find_element_by_css_selector('a').text
        except:
            dic["TenNhaCungCap"] = ""
        try:
            dic["DiaChi"] = temp[7].find_element_by_css_selector('address em').text
        except:
            dic["DiaChi"] = ""        
    except:
        pass
    return dic
```

```python
listThuoc = []
for i in range(len(data)):
    driver.get(data[i])
    temp = getInforOfThuoc()
    print(str(i/len(data) * 100)+ " %")
    if(temp == {}):
        pass
    else:
        listThuoc.append(temp)

with open('dataCrawl.json', 'w') as f:
    json.dump(listThuoc, f)
```
